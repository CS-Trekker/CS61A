## æ‰“å°è·¯å¾„

````python
import os
from pathlib import Path
import emojis

def list_files_with_emoji_recursive(directory, level=0):
    # è·å–è·¯å¾„å¯¹è±¡
    path = Path(directory)

    # æ£€æŸ¥è·¯å¾„æ˜¯å¦å­˜åœ¨
    if not path.exists():
        print("è·¯å¾„ä¸å­˜åœ¨")
        return

    # éå†è·¯å¾„ä¸­çš„æ‰€æœ‰æ–‡ä»¶å’Œç›®å½•
    for item in path.iterdir():
        # ä½¿ç”¨è¡¨æƒ…ç¬¦å·åŒºåˆ†æ–‡ä»¶å’Œç›®å½•
        indent = '\t' * level
        if item.is_dir():
            print(f"{indent}{emojis.encode(':file_folder:')} {item.name}")
            # é€’å½’è°ƒç”¨ä»¥åˆ—å‡ºå­ç›®å½•
            list_files_with_emoji_recursive(item, level + 1)
        else:
            print(f"{indent}{emojis.encode(':page_facing_up:')} {item.name}")

# æŒ‡å®šè·¯å¾„
directory = ".."

# è°ƒç”¨å‡½æ•°æ‰“å°æ–‡ä»¶å’Œç›®å½•
list_files_with_emoji_recursive(directory)
````
